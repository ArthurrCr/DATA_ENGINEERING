# Requirements

Neste ponto, gostaria que você começasse a pensar nos requisitos como uma hierarquia de necessidades. No topo dessa hierarquia, temos os objetivos e metas do negócio. Esses são os elementos que descrevem como o sucesso se parece para o negócio como um todo. Esses objetivos podem envolver o crescimento da receita, participação no mercado, base de usuários ou algo mais que represente um resultado ideal para a empresa.

A próxima camada na hierarquia são as necessidades das partes interessadas. Agora, você me ouvirá falar muito sobre stakeholders (partes interessadas) ao longo destes cursos. Mas, neste caso, você pode entender essas partes interessadas como os funcionários individuais de uma empresa, sendo que cada um deles desempenha um papel para cumprir os objetivos empresariais de alto nível. Para que o trabalho seja feito com sucesso, as partes interessadas têm necessidades. Elas precisam de coisas como recursos adequados, ferramentas e gestão, entre outras. Para os fins desses cursos, você pode assumir que elas também têm uma necessidade de sistemas de dados robustos.

Abaixo das necessidades das partes interessadas nessa hierarquia, temos os requisitos do sistema. Em princípio, isso pode se referir a qualquer sistema, mas aqui estamos falando dos sistemas de dados que você construirá e manterá. Trata-se de um conjunto de requisitos que os seus sistemas de dados precisam atender para servir às necessidades das partes interessadas. Agora, vamos separar os requisitos do sistema em requisitos funcionais e não funcionais.

Requisitos funcionais são aqueles requisitos do sistema que podem ser expressos em termos de uma funcionalidade específica, ou seja, o que o sistema será capaz de fazer para atender às necessidades das partes interessadas. Por exemplo, um requisito funcional para um sistema de dados que monitora transações bancárias em busca de comportamento fraudulento pode ser que o sistema seja capaz de reportar imediatamente transações potencialmente fraudulentas.

Requisitos não funcionais, por outro lado, são aqueles requisitos do sistema que você pode pensar como características ou atributos do sistema que permitem que ele funcione adequadamente. Esses podem ser atributos relacionados a aspectos como latência, escalabilidade, confiabilidade, custo ou segurança. Por exemplo, um requisito não funcional relacionado à escalabilidade para um pipeline de streaming que ingere dados de uma plataforma de e-commerce pode ser: o sistema deve ser capaz de escalar para ingerir dados de 10.000 usuários comprando simultaneamente.

E aí está. Essa é uma hierarquia de necessidades para o negócio, relacionada ao seu trabalho como engenheiro de dados. A principal lição aqui é que, como temos falado até agora, o seu trabalho na construção de sistemas de dados está diretamente conectado ao trabalho de outras pessoas na sua organização e aos objetivos gerais do negócio. Por isso, é tão importante entender as necessidades das suas partes interessadas e os objetivos do negócio para o trabalho que você faz.

Isso nos leva de volta ao levantamento de requisitos. Na primeira semana deste curso, começamos com uma conversa com os cientistas de dados para entender suas necessidades. Você pode pensar nesse exercício como um levantamento de requisitos que começou em algum ponto do meio dessa hierarquia. Idealmente, no entanto, você gostaria de começar no topo. Isso significa conversar com a liderança sobre os objetivos da empresa. Se você trabalhar em uma empresa pequena, pode até ser possível conversar diretamente com o CEO. Eu definitivamente recomendaria isso. Se você estiver em uma organização maior, basta ir o mais alto possível dentro da cadeia de comando para se orientar sobre os objetivos do negócio.

Na primeira semana do curso, você viu uma conversa que tive com Soul, que foi o diretor de dados (CDO) em várias grandes organizações. Agora, nem todas as empresas têm um CDO, mas a maioria terá um diretor de tecnologia (CTO), que é o executivo responsável pela tecnologia da empresa. O CTO é responsável por liderar o departamento de engenharia para usar a tecnologia a fim de melhorar produtos e serviços, com o objetivo de fazer o negócio crescer.

# Breaking Down the Conversation with Marketing

A forma como gosto de documentar os requisitos é utilizando a hierarquia que vimos em um vídeo anterior. Assim, é fácil visualizar as conexões dos requisitos do sistema, que estão na parte inferior, até os objetivos de negócios de alto nível, que estão no topo. Aqui, começo escrevendo os objetivos de negócios conforme eu os entendi. Neste caso, os objetivos de negócios que aprendemos são que a empresa visa continuar em sua trajetória de crescimento, com foco em retenção e lealdade de clientes, além de expansão para novos mercados e novos produtos. Nessas iniciativas, a empresa busca ser orientada por dados em sua tomada de decisões.

Abaixo dos objetivos de negócios, agora podemos adicionar as necessidades das partes interessadas. Por enquanto, vou considerar os cientistas de dados e o gerente de marketing de produtos como as partes interessadas com as quais conversamos até agora. Neste ponto, sabemos que eles precisam de duas coisas principais: alguns painéis de análise e um sistema de recomendação. Vamos primeiro analisar os painéis de análise. Tanto os cientistas de dados quanto o gerente de marketing de produtos disseram que precisam de dados "em tempo real" ou "dados atuais" apresentados nos painéis. Ao investigar um pouco mais e perguntar qual ação eles planejam tomar com os dados, descobrimos que a equipe de marketing gostaria de estar ciente de quando ocorrem picos de demanda para determinados produtos, para que possam reagir com campanhas promocionais adicionais. Dado que a duração desses picos de demanda é esperada para durar de várias horas a um ou dois dias, parece que atualizações de painéis a cada hora seriam suficientes, o que a equipe de marketing confirmou como atendendo bem às suas necessidades. 

Aqui, temos uma necessidade da parte interessada, e você poderia registrar o requisito funcional relacionado para o sistema de dados assim: O sistema de dados precisa fornecer dados transformados que não sejam mais antigos do que uma hora. 

Quando se trata do sistema de recomendação, parece que a solução atual é apenas exibir alguns produtos populares na página de checkout. O que a equipe de marketing gostaria de ter, em vez disso, é um sistema que forneça recomendações de produtos personalizadas com base no histórico de navegação ou compras de um cliente, assim como o que ele tem no carrinho de compras no momento do checkout. Aqui temos outra necessidade da parte interessada e, no que diz respeito a um requisito funcional para o seu sistema de dados para atender a essa necessidade, você poderia escrever algo como: o sistema precisa fornecer os dados de treinamento apropriados para o desenvolvimento do modelo de recomendação. Depois, ele precisa ser capaz de ingerir, transformar e fornecer dados do usuário para o modelo de recomendação treinado e, em seguida, retornar as saídas do modelo, talvez na forma de IDs de produtos, de volta para a plataforma de vendas.

Agora, gostaria de fazer uma pausa por um momento e destacar o fato de que traduzir as necessidades das partes interessadas em requisitos funcionais para o seu sistema pode ser um pouco difícil. Como você pode ver aqui com o painel de análise, o que a equipe de marketing realmente precisa são os próprios painéis, e pode ser tentador escrever detalhes sobre os recursos do painel ou as métricas a serem exibidas. Mas, na verdade, construir o painel será responsabilidade dos cientistas de dados. Para fornecer esses recursos e métricas, os cientistas de dados precisam de acesso a dados que não sejam mais antigos do que uma hora. Quando se trata do seu sistema de dados, enquanto você for capaz de fornecer os dados de forma oportuna para que os cientistas de dados possam trabalhar, seu sistema estará atendendo a esse requisito funcional.

# Documenting Nonfunctional Requirements

Discutimos um exemplo de conversa entre você, o engenheiro de dados, e o engenheiro de software responsável pelo sistema de origem de dados do qual você está planejando ingerir dados. Agora, vamos revisar alguns dos principais aprendizados dessa conversa e continuar a documentação dos requisitos para os nossos sistemas de dados.

Na conversa com o engenheiro de software, aprendemos alguns pontos chave. Primeiro, eles sugeriram que uma forma de resolver o problema de acesso aos dados seria configurar um banco de dados de réplica de leitura e uma API para fornecer acesso contínuo aos dados. Isso é, na verdade, uma abordagem bastante comum para fornecer acesso aos dados, enquanto se protege os sistemas de produção. Eles também discutiram brevemente algumas maneiras de garantir um grau de estabilidade na réplica de leitura, a fim de minimizar as mudanças disruptivas para os sistemas downstream. Agora, vamos dar uma olhada em como documentar os requisitos para os sistemas de dados que estamos tentando construir. **Requisitos Não Funcionais**

Os requisitos não funcionais podem ser um pouco mais complicados do que os requisitos funcionais, no sentido de que, geralmente, não são coisas que seus stakeholders solicitam explicitamente. Em vez disso, eles são as características ou atributos que seu sistema deve ter para que o trabalho seja bem feito.

**Para os Painéis de Análise:**

Já escrevemos o requisito funcional de que os dados devem ser ingeridos, transformados e servidos de tal forma que fiquem disponíveis no banco de dados usado para os painéis de controle, no máximo uma hora após serem gravados no sistema de origem. Agora, podemos escrever um requisito não funcional relacionado à **escalabilidade** e **latência**. Dado que o volume de dados ingeridos pode variar dependendo de quantos usuários estão na plataforma, um requisito não funcional poderia ser que o sistema será capaz de escalar para ingerir, transformar e servir o volume de dados esperado, com um nível máximo de atividade de usuários na plataforma, enquanto se mantém dentro dos requisitos de latência. Em outras palavras, o processamento dos dados não pode desacelerar devido ao aumento do volume de dados.

Outro requisito não funcional para os painéis pode ser relacionado à **confiabilidade** e **manutenibilidade** do sistema. Neste caso, como mencionado na conversa com o engenheiro de software, você poderia configurar verificações nos dados para garantir que eles estejam no formato esperado, e os engenheiros de sistema de origem podem informar quando esperar mudanças no esquema de dados. Em termos de confiabilidade, seu requisito não funcional poderia ser que o sistema execute verificações de qualidade de dados para garantir que os dados ingeridos estejam conformes. E, em termos de manutenibilidade, você poderia escrever que as fases de ingestão e transformação devem ser facilmente adaptáveis para acomodar quaisquer mudanças no esquema dos dados.

**Para o Sistema de Recomendação:**

Aqui, novamente, você teria um requisito de **latência**. Como você pode imaginar, se o objetivo é fornecer recomendações de produtos enquanto alguém está navegando ou comprando itens, o tempo de resposta precisa ser relativamente instantâneo. Até agora, temos falado sobre dados em tempo real ou algo mais próximo disso, mas o que isso realmente significa? Na prática, definir "tempo real" depende das nuances do sistema que você está construindo. Mas aqui, vamos assumir que o sistema de dados deve ter uma latência de menos de um segundo, desde a ingestão dos dados do usuário até a entrega das recomendações. Em outras palavras, o tempo que passa entre o momento em que o sistema recebe dados do cliente da plataforma e serve os dados de recomendação de volta para a plataforma deve ser inferior a um segundo. Pode haver circunstâncias em que o requisito de latência seja maior ou menor, mas um segundo é pelo menos o que seria esperado para um sistema de recomendação típico.

Você também teria um requisito não funcional relacionado à **escalabilidade** para o sistema de recomendação. Semelhante ao aplicativo de painéis, aqui, seria necessário que o seu sistema fosse capaz de escalar até o número máximo de usuários simultâneos na plataforma. E você também poderia ter um requisito não funcional em termos de **confiabilidade**. Por exemplo, se o sistema falhasse ao gerar recomendações ou experimentasse algum outro erro, você gostaria de executar um comportamento previsível que não causasse uma experiência ruim para o usuário. Então, por exemplo, poderíamos capturar o requisito de confiabilidade da seguinte forma: o sistema deve sempre retornar um conjunto de recomendações em até um segundo, de forma que, caso o pipeline de recomendação falhe por qualquer motivo, ele deva voltar a servir uma seleção dos produtos mais populares.

Agora você tem uma série de requisitos não funcionais para esses dois sistemas. Como mencionei antes, na prática, você pode ter muitos requisitos funcionais e não funcionais para qualquer sistema de dados. Aqui, só documentamos alguns dos mais evidentes para esses dois sistemas.

Até agora, nos concentramos em descobrir como construir sistemas de dados que otimizem para atender às necessidades das partes interessadas. Algo que ainda não discutimos esta semana é sobre como avaliar os trade-offs no levantamento de requisitos. Por exemplo, suas partes interessadas podem querer que você construa um sistema de dados o mais rápido possível, ou, quando se trata de custos, você pode estar trabalhando com um orçamento limitado para os sistemas que constrói. Na realidade, claro, restrições de cronograma e orçamento serão parte de qualquer projeto em que você trabalhe. Suas conversas com as partes interessadas precisarão incluir uma discussão sobre o que é mais importante: os recursos do sistema, o cronograma de implantação ou o custo.

Há um conceito que surgiu da gestão de projetos, conhecido como o triângulo de ferro, onde você tem três aspectos de qualquer projeto que estão fundamentalmente em tensão uns com os outros. Esses são o cronograma do projeto, o escopo do trabalho e o custo. O que quero dizer com "em tensão" é que você pode pensar em cada um desses elementos puxando em uma direção diferente. Por exemplo, se você aumentar o escopo de um projeto, então ou o cronograma ou o custo, ou ambos, devem aumentar; ou se você quiser concluir o projeto em um prazo mais curto, isso pode exigir um aumento do custo ou redução do escopo, ou ambos. Há até um ditado antigo que se desenvolveu em torno desse conceito, que é "bom, rápido ou barato. Você só pode escolher dois". Em outras palavras, se você quer algo bom e rápido, então não será barato, ou se você quer algo rápido e barato, então não será bom. E assim por diante.

Na realidade, toda empresa quer que os projetos sejam bem feitos e, normalmente, quer que sejam concluídos o mais rápido possível e dentro de certas restrições orçamentárias. O que você faz? Bem, como muitos autores apontaram desde o surgimento do triângulo de ferro, a ideia de que você não pode otimizar simultaneamente os três aspectos é, na verdade, uma falácia. Eu não vou entrar em detalhes agora, mas recomendaria que você fizesse uma busca sobre a falácia do triângulo de ferro ou algo semelhante e veja o que encontra.

Isso não quer dizer que você possa evitar a necessidade de fazer trade-offs entre custo, escopo e prazos nos sistemas que você constrói. Em resumo, a forma de quebrar o triângulo de ferro é através da aplicação de princípios e processos como os que temos discutido ao longo deste curso, coisas como construir sistemas fracamente acoplados, otimizar para decisões de portas de duas vias e entender profundamente as necessidades das partes interessadas. Ao aplicar esses princípios e processos, você será mais capaz de construir e manter sistemas de dados de alta qualidade dentro de um cronograma e orçamento previsíveis.

# Details of the Recommender System

Dado as características de um usuário específico, o modelo prevê quais avaliações o usuário atribuirá a um produto específico. Ao calcular as avaliações previstas para todos os produtos da empresa, o modelo pode fornecer uma lista de recomendações para o usuário (“Com base no seu perfil pessoal, você pode gostar de…”). O modelo é treinado para calcular dois vetores chamados *embeddings*: um descrevendo as características de um produto (product embedding) e outro descrevendo as preferências do usuário (user embedding). Com base nesses dois embeddings, o modelo pode prever se o usuário gostará de um produto.

Os embeddings de produtos também podem fornecer outro conjunto de recomendações para o usuário. Com base nas interações do usuário com produtos na plataforma da empresa (produtos comprados, adicionados ao carrinho ou pesquisados), produtos semelhantes podem ser encontrados e recomendados ao usuário (“Com base no seu histórico de navegação, você pode gostar de…”). Para qualquer produto com o qual o usuário tenha interagido em uma sessão, o modelo de recomendação calcula o vetor de *embedding* desse produto. Esse vetor de *embedding* é então comparado com os embeddings de todos os produtos da empresa para encontrar itens semelhantes (com base nas suas distâncias). Para encontrar itens semelhantes, você implementará um banco de dados vetorial no laboratório, o que torna a busca por produtos semelhantes mais rápida do que o uso do método de força bruta.

O AWS Lambda foi uma das primeiras e continua sendo uma das ferramentas sem servidor mais populares na AWS. Com funções Lambda, você pode executar código em resposta a um gatilho ou evento e, para o seu pipeline ETL nesta semana, você certamente poderia considerar o uso de uma função Lambda para extrair dados do sistema de origem, aplicar transformações e enviá-los para o armazenamento. No entanto, as funções Lambda têm limitações. Coisas como um tempo limite de 15 minutos para cada chamada de função e limitações quanto à memória e alocação de CPU para cada função, entre outras, o que pode significar que você precisará dividir sua tarefa em partes menores para se manter dentro dessas limitações. Além disso, escrever funções Lambda exige que você escreva código personalizado para seu caso de uso, o que, neste caso, novamente, pode não ser o melhor uso do seu tempo.

Em termos de ferramentas sem servidor especificamente para o processamento de dados em batch, há dois serviços sobre os quais eu gostaria de falar. São eles o Amazon Glue ETL e o Amazon EMR Serverless. De certa forma, você pode dizer que há uma quantidade considerável de sobreposição entre o que esses dois serviços podem fazer. E, na verdade, pode ser que as nuances do seu projeto específico determinem qual ferramenta pode atender melhor às suas necessidades. Mas, em um nível alto, a diferença entre os dois se resume às compensações entre controle e conveniência. Simplificando, o EMR Serverless oferece mais controle sobre o que você pode fazer, enquanto o Glue ETL oferece uma experiência mais conveniente. Mas vamos entrar nos detalhes para que você possa ver o que quero dizer com isso.

O EMR foi projetado como uma ferramenta de big data que oferece suporte a uma ampla gama de frameworks como Apache Spark e Apache Hive. Portanto, se você faz parte de uma equipe que realiza análises de petabytes de dados, talvez usando Hadoop ou exigindo a flexibilidade de incorporar seus próprios componentes personalizados, o EMR ou o EMR Serverless pode ser a escolha certa. O Glue ETL também pode lidar com grandes cargas de dados, mas as verdadeiras vantagens são os recursos adicionais aos quais você tem acesso. Por exemplo, ao se conectar a um sistema de origem, o Glue usa algo chamado crawlers, que automaticamente descobrem e classificam dados, criando metadados no processo, incluindo definições de tabela e esquemas. Esses metadados são então usados para preencher o Glue Data Catalog, que é um repositório central contendo informações sobre todos os seus ativos de dados. Com essas informações sobre seus dados agora no seu catálogo de dados, você pode usar a ferramenta Visual ETL do Glue para projetar seu pipeline usando uma interface gráfica no console de gerenciamento da AWS, que gerará automaticamente o código Spark necessário para rodar seu pipeline. Quando você executa seu pipeline, os servidores mantêm o Glue Data Catalog para rastrear as transformações que você aplica, e esse catálogo pode ser usado a jusante para integrar mais facilmente com outros serviços da AWS.

Existem outras opções que você poderia considerar para construir a ingestão e transformação do seu pipeline ETL na AWS, mas essas são as principais opções que eu queria compartilhar com você.

Quando se trata dos aspectos de armazenamento e serviço do pipeline de batch com o qual você está trabalhando nesta semana, sua escolha dependeria principalmente do caso de uso a jusante que você está atendendo. Por exemplo, se você está ingerindo dados tabulares normalizados, aplicando transformações para modelá-los em um esquema estrela para análises, então uma opção que você poderia escolher seria armazená-los e servi-los em outra instância do RDS. Ou se você quiser executar consultas analíticas complexas em grandes conjuntos de dados e aproveitar outros recursos que os data warehouses oferecem, você poderia escolher armazenar e servir os dados no Amazon Redshift, que é uma solução poderosa de data warehouse, embora a um custo significativamente mais alto que o RDS. Nesta semana, você estará atendendo a um caso de uso de aprendizado de máquina, onde os dados serão usados para treinar um modelo de recomendação. Quando o consumidor de dados a jusante for outro profissional técnico de dados que planeja manipular os dados e incorporá-los em seus próprios sistemas, muitas vezes a melhor e mais barata opção de armazenamento e serviço é o armazenamento de objetos no Amazon S3.

No mundo real, para sistemas de dados baseados na AWS, é relativamente comum que o S3 sirva como uma espécie de área de preparação como esta, porque o S3 é flexível, escalável e relativamente econômico. Ele permite que você armazene praticamente qualquer tipo de dado que se integre facilmente com outros serviços da AWS. No S3, você armazena dados em buckets, que são contêineres para objetos. # AWS Services for Streaming Pipelines

Você aprendeu sobre alguns dos serviços da AWS que você pode escolher ao construir pipelines de processamento em lote. No laboratório desta semana, além de um componente de processamento em lote, seu sistema também terá um componente de streaming. Neste vídeo, vamos explorar os serviços de streaming na AWS. Como você aprendeu com o Joe, os dados de streaming podem vir de diversas fontes. Eles podem ser de dispositivos IoT ou dados de cliques de um site ou aplicativo móvel. Sua fonte de streaming pode até ser um banco de dados, no sentido de que você pode estar transmitindo continuamente quaisquer mudanças ou atualizações nos dados do banco de dados por meio de um processo chamado captura de dados de mudanças (CDC). E assim como para o processamento em lote, uma forma de ingerir dados de streaming seria iniciar uma instância EC2 e escrever alguns scripts personalizados para realizar a CDC ou se conectar a qualquer outra fonte de streaming. Depois, você transformaria os dados e os enviaria para o destino desejado. Assim como no processamento em lote, essa abordagem baseada em EC2 significa que você seria responsável por instalar o software, gerenciar a segurança e toda a complexidade que envolve a implantação de um servidor na nuvem. Seguindo as mesmas ideias que discutimos sobre o processamento em lote, você também poderia considerar as funções Lambda como uma opção serverless para criar seu próprio sistema de streaming. Mas, novamente, isso envolve escrever código personalizado para o que for necessário e as possíveis limitações do Lambda para o seu caso de uso específico. Além disso, interagir com fontes de streaming pode ser mais complexo do que executar cargas de trabalho em lote. Posso dizer que você provavelmente não deve construir suas próprias soluções de streaming personalizadas com EC2 ou Lambda, a menos que tenha 100% de certeza de que seu caso de uso não pode ser resolvido com ferramentas abertas ou gerenciadas já existentes.

Vamos dar uma olhada em alguns dos serviços da AWS que podem suportar cargas de trabalho de streaming. O primeiro é o Amazon Kinesis Data Streams. Este é um serviço popular da AWS que permite a ingestão de dados em tempo real. O funcionamento é o seguinte: você tem produtores de dados que enviam dados para o Kinesis Data Streams, e isso pode ser dados de logs de servidores web, dados de dispositivos IoT ou dados de cliques, entre outros exemplos. O Kinesis em si não é exigente quanto ao tipo de dado que você envia. Ele é agnóstico em relação aos dados, ou seja, você pode enviar JSON, XML, dados estruturados ou não estruturados para o stream de dados. Os dados são enviados para o stream pelos produtores e depois armazenados no Kinesis por um período de tempo configurável. O tempo de retenção padrão e mínimo é de 24 horas, mas também pode ser estendido. Os dados armazenados podem ser recuperados pelos consumidores do stream de dados. Vários consumidores podem puxar os mesmos dados e processá-los de maneiras diferentes. É comum que os consumidores do Kinesis peguem os dados e os coloquem em outro lugar, como um serviço de armazenamento ou um data warehouse, ou que os consumidores realizem alguma análise em tempo real dos dados que estão passando pelo stream. Esses consumidores podem ser aplicativos de software que estão puxando dados do stream e processando-os, e esses aplicativos podem estar rodando em serviços de computação como EC2 ou Lambda.

Além do Kinesis, outra opção que você tem é o Amazon Managed Streaming for Apache Kafka, também conhecido como MSK. O MSK é um serviço que fornece grande parte das mesmas funcionalidades do Kinesis Data Streams. O Apache Kafka é uma plataforma de streaming open-source, popular em diversos casos de uso de streaming. O MSK é um serviço totalmente gerenciado que facilita a construção e a execução de aplicativos que utilizam o Apache Kafka para processar dados de streaming. O MSK roda versões open-source do Kafka, o que é útil porque qualquer aplicativo, ferramenta ou plugin existente da comunidade Apache Kafka é compatível. O funcionamento do MSK é o seguinte: você primeiro cria um cluster Apache Kafka, e o serviço MSK gerencia o pesado trabalho de provisionar e operar os nós que rodam o Kafka para você. Isso permite que você evite o trabalho pesado indesejado e possa focar mais no seu aplicativo personalizado. Depois, você, como usuário, interage com o que é chamado de plano de dados Kafka, que o MSK gerencia para você, criando tópicos, produzindo e consumindo dados. Produtores de dados e consumidores de dados se conectam ao cluster para enviar e receber mensagens. Tanto o Kinesis Data Streams quanto o MSK podem escalar para lidar com volumes de dados em nível de petabytes provenientes de múltiplas fontes de dados com latência de milissegundos, o que possibilita processamento e análise em tempo real.

Quando se trata de escolher um em relação ao outro, assim como ocorre ao escolher entre ferramentas semelhantes para outros aspectos dos seus sistemas de dados, vai depender do seu caso de uso. Mas, de forma geral, você pode pensar nisso como um trade-off entre controle e conveniência. Se você for novato em arquiteturas de dados de streaming, o Kinesis é frequentemente recomendado pela sua facilidade de uso relativa e pela menor sobrecarga operacional. Por outro lado, se você já está executando um cluster Kafka ou talvez tenha experiência técnica com Kafka internamente, ou se estiver buscando um maior grau de flexibilidade e controle, o MSK pode ser a melhor escolha.

Em relação ao caso de uso de ler os dados de um stream e armazená-los em outro local, o próximo serviço que quero mencionar é o Amazon Data Firehose. Para dar um pouco de contexto sobre por que o serviço Data Firehose foi criado, acontece que o Kinesis foi lançado primeiro como um serviço para sistemas de dados de streaming. A AWS percebeu que muitos usuários do Kinesis Data Streams estavam simplesmente pegando os dados do stream e armazenando-os no S3 ou em outro lugar. Mas para trabalhar com o Kinesis dessa maneira, você precisava escrever código personalizado que criasse a conexão com o stream de dados, lesse o stream, fragmentasse os dados e depois os armazenasse. Para facilitar esse processo, a AWS criou o serviço Amazon Data Firehose, que pode ser integrado com o Kinesis Data Streams e é projetado para permitir que você pegue dados de um stream e os armazene em um destino como S3 ou Redshift, ou os envie para endpoints HTTP ou provedores de serviços terceirizados como Data Dog ou Splunk. A principal vantagem do Data Firehose é que ele facilita a leitura dos dados de um stream e o envio para o armazenamento sem a necessidade de escrever código personalizado ou criar integrações complexas. Além de se integrar com o Kinesis Data Streams, o Data Firehose também pode se integrar com mais de 20 outras fontes de dados de ADS para ingerir dados de streaming, incluindo MSK e outros.
