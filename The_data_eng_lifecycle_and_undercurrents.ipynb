Os sistemas de origem mais comuns so os bancos de dados, que podem ser bancos de dados relacionais, representados como tabelas de dados relacionados, ou outros tipos de sistemas NoSQL, incluindo bancos de dados de chave-valor, repositrios de documentos e outros. ### Desafios com Sistemas de Origem

Em um mundo ideal, os sistemas de origem dos quais voc depende forneceriam os dados necessrios de maneira consistente e pontual, permitindo que voc construsse sistemas downstream que se baseassem na previsibilidade desse sistema de origem. Esses sistemas, s vezes, ficam fora do ar, a equipe que gerencia o sistema pode alterar o formato ou o esquema dos dados, ou talvez o esquema permanea o mesmo, mas os dados em si mudem.

Quando comecei a trabalhar como engenheiro de dados, lembro-me de ter trabalhado com um banco de dados mantido por uma equipe interna de engenheiros de software. ### Importncia do Entendimento dos Sistemas de Origem

Ao acessar dados de sistemas de origem,  essencial entender como esses sistemas esto configurados e quais mudanas nos dados ou no sistema voc pode esperar. Isso significa que, como engenheiro de dados, voc ter mais sucesso se trabalhar diretamente com os responsveis pelos sistemas de origem para compreender como esses sistemas funcionam, como eles geram dados, como esses dados podem estar sujeitos a mudanas ao longo do tempo e, em ltima instncia, como essas mudanas afetaro os sistemas downstream que voc constri. # Ingestion

**Como engenheiro de dados, voc construir arquiteturas que comeam com o passo de ingesto de dados a partir dos sistemas de origem, o que significa mover dados brutos dos sistemas de origem para o seu pipeline de dados para processamento posterior.**  
Na minha experincia, os sistemas de origem e a ingesto de dados representam os maiores gargalos no ciclo de vida da engenharia de dados.

**Entendimento dos Sistemas de Origem**  
Se voc comeou trabalhando diretamente com os responsveis pelos sistemas de origem para entender como esses sistemas funcionam, como eles geram dados, de que maneira esses dados podem estar sujeitos a alteraes ao longo do tempo e, por fim, como essas mudanas impactaro os sistemas a jusante que voc construir, ento estar bem posicionado para evitar armadilhas comuns na fase de ingesto.

**Deciso Crtica: Frequncia da Ingesto de Dados**  
Uma das decises crticas que voc precisa tomar ao projetar processos de ingesto de dados  com que frequncia voc precisa ingerir os dados, ou seja, com que regularidade mover os dados desses sistemas de origem  mostrados  esquerda  para o seu pipeline de dados para processamento posterior.  
Voc pode optar por ingerir dados periodicamente em lotes, como uma vez por hora ou uma vez ao dia. Outra abordagem, que vem se tornando cada vez mais comum,  ingerir dados como um fluxo constante de eventos em quase tempo real.

**Padres de Ingesto: Lote versus Streaming**  
Focarei na introduo desses dois principais padres de ingesto de dados, lote versus streaming.



- **Ingesto em Lote:**  
  Voc pode pensar nos dados produzidos como uma srie contnua de eventos  esses eventos podem ser cliques em um site, medies de sensores ou outros. No mundo, eventos como esses acontecem continuamente. Em certo sentido, quase todos os dados podem ser vistos como sendo transmitidos em tempo real em sua origem. A ingesto em lote  apenas uma maneira conveniente de processar esse fluxo de eventos em grandes blocos, seja em intervalos de tempo predeterminados ou  medida que os dados atingem um limite de tamanho pr-estabelecido. 
  
  
  
  Por exemplo, voc pode processar todos os dados de um dia inteiro em um nico lote. Por muito tempo, o processamento em lote foi a forma padro de ingerir dados. Hoje, existem mais opes para a ingesto, mas o processamento em lote continua sendo uma abordagem prtica e popular, especialmente em casos onde os dados so usados para anlises e aprendizado de mquina.

- **Ingesto via Streaming:**  

  

  Com a ingesto em streaming, por outro lado, voc est ingerindo e fornecendo dados aos sistemas a jusante de forma contnua e quase em tempo real. Quando digo que est ingerindo dados em quase tempo real, quero dizer que est disponibilizando esses dados para os sistemas a jusante em um curto perodo aps sua produo  possivelmente em menos de um segundo. Nesse caso,  necessrio usar ferramentas especficas, como plataformas de streaming de eventos ou filas de mensagens, para ingerir continuamente os fluxos de eventos. Com essas ferramentas se tornando mais onipresentes, a ingesto via streaming tornou-se mais acessvel e popular.


**Consideraes e Trade-offs na Escolha da Abordagem**  
No entanto, a ingesto via streaming no  necessariamente a melhor escolha para todos os casos de uso. Existem trade-offs significativos a serem considerados ao decidir se a ingesto via streaming  a opo apropriada em vez da abordagem mais simples de ingesto em lote. Para trabalhar em uma soluo de processamento de fluxo, voc deve se perguntar:

- Quais aes podem ser realizadas com os dados em tempo real que representem uma melhoria em relao aos dados processados em lote?
- A ingesto via streaming custa mais do que a ingesto em lote em termos de tempo, dinheiro, manuteno e possveis perodos de inatividade?
- Como a escolha entre ingesto via streaming e ingesto em lote influenciar o restante do seu pipeline de dados?

A ingesto em lote  uma excelente abordagem para muitos casos de uso comuns, como negociaes de modelos e relatrios semanais. Meu conselho  adotar um sistema de ingesto via streaming somente depois de identificar um caso de uso de negcio que justifique os trade-offs de utiliz-lo em comparao ao lote.

**Coexistncia de Ingesto via Streaming e Lote**  
 importante notar que a ingesto via streaming geralmente coexiste com o processamento em lote. Por exemplo, modelos de aprendizado de mquina so geralmente treinados com base em processamento em lote. Contudo, esses mesmos dados de treinamento podem ter sido originalmente ingeridos via streaming devido a algum objetivo separado da sua arquitetura, como a deteco de anomalias em tempo real. Raramente os engenheiros de dados tm a opo de construir um pipeline de dados puramente em streaming sem componentes de lote. Em vez disso, voc escolher onde ocorrero as divises entre lote e streaming.

**Outras Nuances na Ingesto de Dados**  
Alm de decidir entre uma abordagem de lote versus streaming, existem outras nuances importantes na etapa de ingesto, tais como:

- **Uso de Change Data Capture (CDC):**  
  Utilizar CDC para acionar certos processos de ingesto baseados nas alteraes dos dados no sistema de origem.
  
- **Abordagem Push versus Pull:**  
  Decidir se o sistema de origem enviar os dados para voc (push) ou se voc os buscar ativamente a partir da origem (pull).

# Storage

**Interagindo com Sistemas de Armazenamento de Dados no Dia a Dia**

Gostaria que voc refletisse por um momento sobre as diversas formas de interao que tem com os sistemas de armazenamento de dados diariamente. Com o seu smartphone, da mesma forma, voc pode estar enviando ou recebendo mensagens ou interagindo com aplicativos, efetivamente movendo dados entre os diversos componentes de armazenamento do seu dispositivo, bem como na nuvem. Quase toda ao que voc realiza em um dispositivo digital ou online envolve a interao com sistemas de armazenamento de dados, e voc pode se deparar com os limites desses sistemas, como ficar sem espao para armazenar fotos no seu telefone ou tentar enviar arquivos que so muito grandes. A funo, o desempenho e as limitaes que voc encontra tero muito a ver com a forma como esses sistemas esto configurados.

**Hardware Bsico de Armazenamento**



Da mesma forma, no seu trabalho como engenheiro de dados, a funo, o desempenho e as limitaes dos sistemas que voc constri tero muito a ver com as solues de armazenamento que voc escolhe para suportar esses sistemas. Para comear, vamos dar uma olhada em alguns dos ingredientes fsicos brutos do armazenamento. No seu dia a dia, voc provavelmente j se familiarizou com diversas formas de armazenamento em estado slido, como cartes de memria flash ou drives de estado slido do seu laptop ou smartphone. Em contrapartida,  provvel que voc no tenha lido dados de um disquete recentemente.

Pode-se perdoar o pensamento de que o mundo deixou para trs os discos magnticos como soluo de armazenamento de dados. No entanto, a verdade  que os discos magnticos ainda formam a espinha dorsal dos sistemas modernos de armazenamento, e isso se deve principalmente ao baixo custo do armazenamento em disco. Atualmente (2024), o armazenamento em disco  cerca de 2 a 3 vezes mais barato que o armazenamento em estado slido.

A RAM, comumente referida como memria,  outra forma de armazenamento fsico com a qual voc pode estar familiarizado. A RAM oferece velocidades de leitura e escrita muito mais rpidas do que os drives de estado slido ou discos, tornando-se um componente crtico de muitas aplicaes e arquiteturas. Contudo, o armazenamento em RAM pode ser de 30 a 50 vezes mais caro que o armazenamento em estado slido. Alm disso, a RAM  voltil no sentido de que, se o seu sistema perder energia, na maioria das vezes, dependendo do tipo de RAM que voc utiliza, a memria  perdida instantaneamente.

Em muitas arquiteturas modernas, os dados passam por armazenamento magntico, drives de estado slido e memria enquanto percorrem as vrias fases de processamento de um pipeline. No entanto, os componentes fsicos de armazenamento so apenas um aspecto de como o armazenamento de dados  implementado em toda a sua arquitetura.

**Sistemas Modernos e em Nuvem de Armazenamento de Dados**

Os sistemas modernos e em nuvem de armazenamento de dados so comumente distribudos por mltiplos clusters e data centers. Isso significa que itens como rede, CPU, serializao, compresso e cache so todos ingredientes brutos crticos para armazenar dados em sistemas de dados modernos.

Como engenheiro de dados, tipicamente voc no ser responsvel por gerenciar os detalhes granulares de como os seus dados se movem e so armazenados em uma rede de data centers e dispositivos fsicos de armazenamento. Em vez disso, voc trabalhar com sistemas de armazenamento como sistemas de gerenciamento de banco de dados ou plataformas de armazenamento de objetos, como o Amazon S3. Dependendo dos requisitos da sua arquitetura, voc tambm pode estar trabalhando com sistemas como Apache Iceberg ou Hoody, sistemas de armazenamento baseados em cache e memria, ou mesmo armazenamento para streaming. Todos esses sistemas de armazenamento de dados so construdos sobre os ingredientes fsicos e outros ingredientes brutos que existem dentro de servidores e clusters, permitindo que esses sistemas ingiram e recuperem dados usando diferentes protocolos de acesso.

**Abstraes de Armazenamento**

Por fim, no seu trabalho como engenheiro de dados,  provvel que voc no trabalhe apenas com sistemas individuais de armazenamento, mas com combinaes de sistemas organizados em abstraes de armazenamento, como um data warehouse, um data lake ou, mais recentemente, a combinao desses conceitos: o data lake house. Com as ferramentas de abstrao de armazenamento, em vez de se preocupar com os detalhes de como os componentes subjacentes esto organizados, voc escolher vrios parmetros de configurao que permitem atender aos requisitos do seu sistema em termos de latncia, escalabilidade e custo.



Se voc pensar no armazenamento como uma hierarquia, como organizaria os diferentes aspectos do armazenamento? No nvel mais bsico, esto os ingredientes brutos do armazenamento de dados, incluindo os diversos componentes fsicos, como disco, RAM e armazenamento em estado slido, bem como os diversos ingredientes brutos no fsicos, como rede e serializao. Em seguida, sobre esses, existem os sistemas de armazenamento que so construdos a partir desses ingredientes brutos, os quais incluem sistemas de banco de dados, armazenamento de objetos e muito mais. Finalmente, no topo da hierarquia, encontram-se as abstraes de armazenamento, que so combinaes de sistemas de armazenamento que permitem atingir as necessidades de armazenamento de dados em um nvel mais alto, sem se preocupar tanto com os detalhes de baixo nvel.

Como engenheiro de dados,  perfeitamente possvel que voc passe grande parte do seu tempo operando no nvel superior dessa hierarquia, o que significa que voc no precisar se preocupar com os detalhes de como exatamente os seus dados esto se movendo entre os diferentes componentes e sistemas de armazenamento. No entanto, voc ser mais eficaz no seu trabalho se dedicar um tempo para entender o funcionamento interno, as capacidades e as limitaes de toda a soluo de armazenamento, at mesmo nos seus ingredientes brutos.

A verdade  que muitos engenheiros de dados atualmente no entendem profundamente os detalhes dos sistemas de armazenamento que constroem. Isso leva a consequncias infelizes quando se trata de desempenho e custo. Por exemplo, uma vez, eu estava prestando consultoria para uma equipe que falhou em considerar esses detalhes. Eles precisavam mover um grande conjunto de dados para o data warehouse e, sem saber, escolheram uma abordagem de inserir diretamente cada linha de dados individualmente, ou seja, estavam ingerindo e escrevendo uma linha de cada vez no data warehouse. Isso no s se revelou muito lento, como tambm muito caro, pois inseres diretas so geralmente cobradas por uso. No final, eles descobriram o erro e migraram para uma abordagem de ingesto em massa, mas somente depois de terem desperdiado um tempo significativo e consumido metade do oramento anual para o data warehouse em apenas uma semana.

# Queries, Modeling, and Transformation

**A etapa de transformao no ciclo de vida da engenharia de dados  realmente onde voc, como engenheiro de dados, comea a agregar valor.**  
Isso ocorre porque a etapa que vem antes da transformao, a saber, a ingesto e o armazenamento de dados brutos provenientes dos sistemas de origem, no agrega valor direto para as partes interessadas a jusante. Como mencionado anteriormente, em seu papel de engenheiro de dados, o panorama geral  que voc obtm dados brutos de algum lugar, os transforma em algo til e, em seguida, os disponibiliza para os usurios finais. A transformao  a etapa em que esses dados se tornam teis.

**Exemplificando o que significa "til":**  
Imagine, por exemplo, um analista de negcios como seu usurio a jusante. Suponha que ele tenha a tarefa de reportar as vendas dirias de uma variedade de produtos. Esse analista pode precisar de informaes como IDs de clientes, nomes de produtos, preos, quantidades, horrios de venda, entre outros. Embora os analistas de dados geralmente dominem SQL, eles contaro com voc para transformar os dados brutos e fornec-los em um formato que seja rpido e fcil de consultar.

**Outro exemplo: cientistas de dados e engenheiros de machine learning**  
Imagine que o seu usurio a jusante seja um cientista de dados ou um engenheiro de machine learning. Alm do SQL, eles podem at ser fluentes em vrias abordagens potenciais para transformao de dados, mas sua funo principal  realmente utilizar os dados para anlises preditivas. Voc pode agregar um valor tremendo ao lidar com a transformao dos dados em estruturas e caractersticas que podem ser usadas diretamente para o treinamento de seus modelos ou anlises.

**Diviso da etapa de transformao:**  
Chamamos essa parte do ciclo de vida da engenharia de dados de transformao. Na realidade, essa etapa  composta por trs partes: consultas, modelagem e transformao. Incluo consultas e modelagem como separadas da transformao, pois esses so componentes crticos de qualquer pipeline de dados que realmente agregam valor quando executados corretamente e apresentam riscos quando feitos de forma inadequada.

**Consultas:**  
Para ilustrar, vamos comear com as consultas. Quando voc consulta dados, est emitindo um pedido para ler registros de um banco de dados ou outros sistemas de armazenamento. Por exemplo, pode ser necessrio consultar dados tabulares e semi-estruturados que esto armazenados em um data warehouse na nuvem. Existem muitas linguagens que voc pode usar para consultar dados, mas focaremos na linguagem de consulta estruturada, ou SQL, que continua sendo uma linguagem de consulta popular e universal.  
Sua consulta pode envolver a limpeza, juno e agregao de dados em vrios conjuntos de dados. Voc pode usar expresses SQL para filtrar os dados de modo que apenas registros especficos sejam recuperados.  

H mais de uma maneira de escrever uma consulta, e consultas mal escritas podem ter consequncias negativas, como impactar a performance de um banco de dados de origem ou causar uma situao conhecida como exploso de linhas, onde uma consulta que inclui uma juno entre tabelas produz muitos mais registros do que o esperado, o que pode comprometer sua infraestrutura de armazenamento. Em outras circunstncias, consultas mal escritas podem simplesmente ser lentas ou abrangentes, causando atrasos a jusante em relatrios ou anlises.  
Na prtica, acontece que a maioria dos engenheiros de dados sabe ler e escrever SQL, mas no esto familiarizados com o funcionamento interno das consultas. Isso pode ter consequncias imprevistas nas arquiteturas que eles constroem.

**Modelagem de Dados:**  
O prximo ponto a ser abordado  a modelagem de dados. Um modelo de dados representa a forma como os dados se relacionam com o mundo real. A modelagem de dados envolve a escolha deliberada de uma estrutura coerente para os seus dados e  um passo crucial para tornar os dados teis para o negcio.  
Por exemplo, voltando ao caso do analista de negcios que precisa criar relatrios de vendas de produtos, pode ser que voc tenha ingerido dados "normalizados" de um banco de dados relacional de origem que contm tabelas separadas para informaes de produtos, detalhes de pedidos, informaes de clientes, etc. Esses dados frequentemente possuem relaes complexas entre si.  
Para atender s necessidades desse analista, pode ser necessrio realizar o que chamamos de desnormalizao desses dados, modelando-os de uma maneira que permita aos analistas consultar e obter rapidamente os dados necessrios para os relatrios.  
Um bom modelo de dados  projetado para refletir da melhor forma possvel os processos, definies, fluxos de trabalho e lgicas da sua organizao. Por exemplo, o termo "cliente" pode ter significados diferentes para diferentes departamentos da sua empresa. Para ter sucesso na modelagem de dados, voc precisa trabalhar com as partes interessadas para entender sua terminologia  como o que o termo "cliente" significa para eles  bem como os objetivos de negcio para os dados.  

**Transformao dos Dados:**  
Alm de consultar e modelar os dados, estes tambm precisam ser transformados, isto , manipulados, aprimorados e preparados para o uso a jusante.  
Como mencionei anteriormente, voc normalmente transformar os dados vrias vezes ao longo do ciclo de vida da engenharia de dados.  
Por exemplo, os dados podem ser transformados antes mesmo de serem acessados por voc, como a adio de um carimbo de data/hora a um registro enquanto ele ainda est em um sistema de origem, ou voc pode aplicar transformaes enquanto os dados esto em trnsito durante a ingesto; logo aps a ingesto, pode-se configurar transformaes bsicas para mapear os dados para os tipos corretos e colocar os registros em formatos padronizados.  
Voc pode enriquecer um registro dentro de um pipeline de streaming com campos adicionais e clculos antes que ele seja enviado para um data warehouse.  
Ainda mais a jusante, pode ser necessrio transformar o esquema dos dados e aplicar desnormalizao, agregaes em larga escala para relatrios ou criar dados com caractersticas especficas (featurizao) para o treinamento de modelos de machine learning.

### Anlises

**Definio e Objetivos:**  
Anlises envolvem o processo de identificar insights e padres chave dentro dos dados. Como engenheiro de dados,  quase certo que voc servir dados que alimentam uma ou mais das trs formas mais comuns de anlises:
- **Business Intelligence (BI)**
- **Anlises Operacionais**
- **Anlises Incorporadas (Embedded Analytics)**

**Business Intelligence (BI):**  
O BI consiste na explorao de dados histricos e atuais da empresa para a descoberta de insights. Nesse contexto, o papel do engenheiro de dados  fornecer dados que, por fim, so apresentados na forma de relatrios ou dashboards, auxiliando os usurios a tomar decises estratgicas e acionveis.  
- **Exemplo:** Analistas das equipes de vendas e marketing podem usar os relatrios e dashboards de BI para identificar padres e tendncias, bem como para monitorar mtricas como engajamento em campanhas de marketing, vendas por regio e ndices de experincia do cliente.  
- **Exemplo Adicional:** Imagine um cenrio em que um analista percebe um aumento repentino nas devolues de produtos. Esse analista pode investigar os relatrios ou dashboards existentes para comparar o fenmeno com tendncias histricas, ou ainda executar consultas SQL no banco de dados que voc forneceu, ou solicitar que dados adicionais sejam servidos para anlises ad hoc.

**Anlises Operacionais:**

O agendamento puro tem sido uma abordagem amplamente utilizada historicamente para vrias tarefas de processamento de dados. No entanto, podem surgir problemas com essa abordagem.
